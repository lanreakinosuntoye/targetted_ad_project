```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns # data visualization library  
import matplotlib.pyplot as plt

%matplotlib inline 
import warnings                        # To ignore any warnings warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('preproccessed_data3.csv')
data
```

```python
data.columns
```

```python
list = ['xtra_data_talk_rev','lr_ne', 'lr_nw', 'lr_unspecified',
       'dm_apple', 'dm_infinix', 'dm_itel', 'dm_nokia', 'dm_other',
       'dm_samsung','os_android', 'os_unspecified','os_proprietory',
        'gender_f', 'gender_m',
       'gender_unspec', 'dt_featurephone', 'dt_smartphone',
       'dt_unspecified']
data = data.drop(list,axis = 1 )
data
```

```python
X = data.drop('event_type',1)
y = data.loc[:,'event_type']
```

```python
# import module
from sklearn.model_selection import train_test_split
# split in train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
#x_train, x_test, y_train, y_test = train_test_split(x_1, y, test_size=0.25, random_state=42)
```

```python
print(X_train.shape,X_test.shape, y_train.shape, y_test.shape)
```

#### Random Forest

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score,confusion_matrix,recall_score,precision_score,accuracy_score
from sklearn.metrics import accuracy_score

# split data train 70 % and test 30 %
#x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)


clf_rf = RandomForestClassifier(random_state=43)      
clr_rf = clf_rf.fit(X_train,y_train)

ac = accuracy_score(y_test,clf_rf.predict(X_test))
print('Accuracy is: ',ac)
cm = confusion_matrix(y_test,clf_rf.predict(X_test))
sns.heatmap(cm,annot=True,fmt="d")

print("Precision: %.3f" % precision_score(y_test,clf_rf.predict(X_test)))
print("Recall: %.3f" % recall_score(y_test,clf_rf.predict(X_test)))
```

#### Feature Extraction
We will use principle component analysis (PCA) for feature extraction. Before PCA, we need to normalize data for better performance of PCA.

```python
# import module
from sklearn.model_selection import train_test_split
# split in train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
#x_train, x_test, y_train, y_test = train_test_split(x_1, y, test_size=0.25, random_state=42)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

#### PCA - Principal Component Analysis

```python
# Apply PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
explained_variance = pca.explained_variance_ratio_
explained_variance
```
